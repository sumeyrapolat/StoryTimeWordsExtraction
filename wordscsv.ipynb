```python
from bs4 import BeautifulSoup;
import pandas as pd
```

```python
# HTML dosya yolunu belirle
file_path = '/kaggle/input/wordsbook/words.html'

# HTML dosyasını aç ve içeriğini oku
with open(file_path, 'r', encoding='utf-8') as file:
    html_content = file.read()

# BeautifulSoup ile HTML'i parse et
soup = BeautifulSoup(html_content, 'html.parser')

# Kelimelerin bulunduğu div'leri bul (örneğin, 'ws1b' sınıfı ile eşleşenler)
word_divs = soup.find_all('div', class_='ws1b')

# Kelimeleri saklamak için bir list kullan
words_list = []

# Her bir div'den kelimeleri al ve listeye ekle, tekrarları önlemek için kontrol et
for div in word_divs:
    word = div.text.strip()
    if word not in words_list:  # Eğer kelime zaten listede yoksa ekle
        words_list.append(word)

# İlgili kelimeleri belirli konumlara ekleme
if "flexible" in words_list:
    index = words_list.index("flexible")
    words_list.insert(index + 1, "evidence")

if "bait" in words_list:
    index = words_list.index("bait")
    words_list.insert(index + 1, "insist")

if "maintain" in words_list:
    index = words_list.index("maintain")
    words_list.insert(index + 1, "snub")

# Listeyi yazdır
for i, word in enumerate(words_list, start=1):
    print(f"Kelime {i}: {word}")
```

```python
# Kelimeleri 12 sütunluk gruplara ayır
columns = [f"Word{i+1}" for i in range(12)]
rows = [words_list[i:i+12] for i in range(0, len(words_list), 12)]

# DataFrame oluştur
df = pd.DataFrame(rows, columns=columns)

# CSV olarak kaydet
df.to_csv("extracted_words.csv", index=False)

print("Kelimeler başarıyla extracted_words.csv dosyasına kaydedildi.")
```
